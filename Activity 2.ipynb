```python
import pandas as pd
data=pd.read_csv('downloads/Data science/diabetes_model.csv')
data
```

```python
#count plot of outcome
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Outcome',data=data)
plt.title('Countplot of Outcome')
plt.show()
```

```python
#effect of age on outcome
sns.boxplot(x='Outcome',y='Age',data=data)
plt.title('Effect of age on outcome')
```

```python
#heatmap on correlation of features in data
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```

```python
#chance of getting diabetes using diabetes pedigree function
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
```

```python
#relationship between insulin and glucose and group by outcome
sns.scatterplot(x='Insulin',y='Glucose',hue='Outcome',data=data)
plt.show()
```
